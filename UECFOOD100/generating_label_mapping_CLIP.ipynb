# CLIP-based Label Mapping

```python
import torch
import clip
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load CLIP model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# Your ingredient label sets
nutrition5k_ingredients = list(nutrition5k_ingredient_level_nutrients['ingr_name'].unique())
foodseg103_ingredients = list(uecfood100_category_names['name'].unique())

# --- Step 1: Embed text labels using CLIP ---
def get_text_embeddings(labels):
    # Remove None, NaN, float, or bad types and convert to string
    clean_labels = [str(label).strip() for label in labels if isinstance(label, str) or not pd.isna(label)]
    
    with torch.no_grad():
        text_inputs = clip.tokenize(clean_labels).to(device)
        text_features = model.encode_text(text_inputs)
        text_features /= text_features.norm(dim=-1, keepdim=True)

    return text_features.cpu().numpy(), clean_labels

# --- Step 2: Get embeddings ---
nutrition5k_embeddings, clean_nutrition5k = get_text_embeddings(nutrition5k_ingredients)
foodseg103_embeddings, clean_foodseg103 = get_text_embeddings(foodseg103_ingredients)

# --- Step 3: Compute cosine similarity ---
similarity = cosine_similarity(nutrition5k_embeddings, foodseg103_embeddings)

# --- Step 4: Find best match for each ingredient ---
mapping = {}
for i, n5k_label in enumerate(clean_nutrition5k):
    best_match_idx = similarity[i].argmax()
    best_match_label = clean_foodseg103[best_match_idx]
    mapping[n5k_label] = best_match_label

# --- Step 5: Display the mappings ---
print("CLIP-based Label Mapping:\n")
for k, v in mapping.items():
    print(f"{k} --> {v}")
```

# Use Natural Language Prompts
An improved version of the code using prompts like "a photo of [ingredient]", which significantly boosts semantic matching:

```python
import torch
import clip
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Load CLIP model
device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)

# Your ingredient label sets
nutrition5k_ingredients = list(nutrition5k_ingredient_level_nutrients['ingr_name'].unique())
foodseg103_ingredients = list(uecfood100_category_names['name'].unique())

# Natural language prompt templates
def to_prompt(label):
    return f"a photo of {label}"

# Embed labels using CLIP
def get_text_embeddings(labels):
    prompts = [to_prompt(label) for label in labels]
    with torch.no_grad():
        text_inputs = clip.tokenize(prompts).to(device)
        text_features = model.encode_text(text_inputs)
        text_features /= text_features.norm(dim=-1, keepdim=True)
    return text_features.cpu().numpy()

# Compute embeddings
nutrition5k_embeddings = get_text_embeddings(nutrition5k_ingredients)
foodseg103_embeddings = get_text_embeddings(foodseg103_ingredients)

# Compute cosine similarities
similarity = cosine_similarity(nutrition5k_embeddings, foodseg103_embeddings)

# Build best match dictionary
mapping = {}
for i, n5k_label in enumerate(nutrition5k_ingredients):
    best_match_idx = similarity[i].argmax()
    best_match_label = foodseg103_ingredients[best_match_idx]
    mapping[n5k_label] = best_match_label

# Show mapping
print("Improved CLIP-based Label Mapping:\n")
for k, v in mapping.items():
    print(f"{k} --> {v}")
```
